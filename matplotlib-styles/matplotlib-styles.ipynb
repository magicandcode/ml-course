# Default Matplotlib Plot Styles

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Get Default Styles
Get `tuple` with default style names and ordinal when sorted.
Sort available styles on style name in ascending order, excluding `'_test_classic'`. Excluding the first style means ordinals start on `1`.

```python
styles_data = [
    (i, style)
    for i, style
    in enumerate(
        sorted(plt.style.available,
               key=lambda style: style.lower()))][1:]
styles_data
```

### Calculate Figure Count
Due to memory constraints, we want to plot max 20 figures in the same cell. While 20 figures is practical in that it only requires calling the plotting function twice, it's inconvenient when running the cells. Therefore it's preferable to section `styles_data` into `n` sized chunks, where each function call will generate `n` or less plots.

```python
styles_count = len(styles_data)
chunks_count = 3
fig_count = styles_count//chunks_count + styles_count%chunks_count
fig_count
```

### Divide Styles Data Into Sub Lists 

```python
n = fig_count  # Rename for convenience in comprehension
styles = [styles_data[i*n:i*n+n]
          for i, style in enumerate(styles_data[::n])]
styles
```

```python
x = np.random.randn(10, 4)
df = pd.DataFrame(x, columns=list('abcd'))
```

```python
from typing import List, Tuple
```

```python
def display_plot_styles(styles: List[Tuple[int, str]]):
    for i, style in styles:
        plt.style.use(style)
        ax = df.plot.bar()
        ax.set(title=f'{i+1}. {style}',
               xlabel=f'Row number ({style})',
               ylabel=f'Random number ({style})')
        ax.legend().set_visible(True)
```

```python
def display_plot_styles_line(styles: List[Tuple[int, str]]):
    for i, style in styles:
        plt.style.use(style)
        ax = df.plot()
        ax.set(title=f'{i+1}. {style}',
               xlabel=f'Row number ({style})',
               ylabel=f'Random number ({style})')
        ax.legend().set_visible(True)
```

```python
def display_plot_styles_pie(styles: List[Tuple[int, str]]):
    for i, style in styles:
        plt.style.use(style)
        fig, ax = plt.subplots()
        ax.pie([1, 4, 3, 5],
               #explode=[1, 0, 0, 0],
               labels=['One', 'Four', 'Three', 'Five'],
               autopct='%1.1f%%',
               #shadow=True,
               startangle=90)
        ax.axis('equal')
        ax.set(title=f'{i+1}. {style}',)
        ax.legend().set_visible(True)
        plt.show();
```

## Bar Graphs

```python
display_plot_styles(styles[0])
```
